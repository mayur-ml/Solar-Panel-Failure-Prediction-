```python
# Load the data into a Pandas DataFrame
df = pd.read_csv("allSolardata.csv")
```

```python
df
```

```python
# shuffling dataframe : Random sampling :
# The fraction of rows and columns to be selected can be specified in the frac parameter. frac=1 means 100%.
df = df.sample(frac = 1 ,ignore_index= True)
```

```python
df
```

```python
# Labeling F and NF 
df = df.replace({"F2M" :"F","F6L": "F","F6M" :"F","F7L":"F","F7M" :"F","F4L" :"F","F4M": "F" ,"F5M":"F",
                 "F0L":"NF" ,"F5L":"F", "F2L":"F","F0M":"NF" ,"F1M":"F", "F1L":"F", "F3L":"F","F3M":"F"})
```

```python
#finding the count of F and NF
df["Label"].value_counts()
```

```python
#checking for null values:
df.isna().sum()
```

```python
# Get the all faulty and non-faulty rows 
faulty = df[df['Label'] == 'F']
non_faulty = df[df['Label'] == 'NF']
```

```python
# shuffling dataframe : Random sampling :
faultynew = faulty.sample(n=10250)
non_faulty_new = non_faulty.sample(n=10250)
```

```python
faultynew
```

```python
non_faulty_new
```

```python
#concating two dataframe
df_concat = pd.concat([faultynew,non_faulty_new] ,ignore_index= True )
```

```python
# shuffling dataframe : Random sampling :
df_concat = df_concat.sample(frac = 1 ,ignore_index= True)
```

```python
df_concat
```

```python
# Chicking unique values
df_concat["Label"].value_counts()
```

```python
# takin 20000 rows dataset for building model
df_new = df_concat.loc[:20000 ,:]

# hold dataset > to check how our model working with new values after deployment
df_hold =df_concat.loc[20000:,:]
```
